```python
from pathlib import Path
import os

# Create output directory
output_dir = Path("../kb/ingested")
output_dir.mkdir(parents=True, exist_ok=True)

# Ingest both generated and manual KB
sources = [
    ("Generated", Path("../kb/generated")),
    ("Manual", Path("../kb/manual"))
]

total_files = 0

for source_name, source_dir in sources:
    if source_dir.exists():
        print(f"\n{'='*50}")
        print(f"Processing {source_name} KB from: {source_dir}")
        print(f"{'='*50}")
        
        file_count = 0
        
        for file_path in source_dir.glob("*.md"):
            try:
                # Read markdown file
                with open(file_path, "r", encoding="utf-8") as f:
                    text = f.read()
                
                # Define output path with source prefix
                output_path = output_dir / f"{source_name.lower()}_{file_path.stem}.md"
                
                # Write to output
                with open(output_path, "w", encoding="utf-8") as f:
                    f.write(text)
                
                print(f"  ✓ {file_path.name} → {output_path.name}")
                file_count += 1
                total_files += 1
            
            except Exception as e:
                print(f"  ✗ Error processing {file_path.name}: {e}")
        
        print(f"  {file_count} files processed from {source_name}")
    else:
        print(f"⚠ Source directory not found: {source_dir}")

print(f"\n{'='*50}")
print(f"Ingestion Complete!")
print(f"Total files ingested: {total_files}")
print(f"Output directory: {output_dir}")
print(f"{'='*50}")
```
